```python
# NIMS236
prior_from_ki = {'n_e': 1076971.567,
                 'n_e_std': 378922.7567662638,
                 'k': 18.20701391029358,
                 'k_std': 3.6527518624450086,
                 'one_t_s': 1.0401270236968994,
                 'one_t_s_std': 0.016833790482064,
                 's_d': 423.2007942504883,
                 's_d_std': 9.446916499964738}

def sn_para_cal(df):
    for i in range(len(df)):
        print(i)
        curve_optimizer = WoehlerCurve_optimizer(fkm=0, calc_posterior=False, prior_from_ki={})
        string_data = df['FT_results'][i]
        dic = json.loads(string_data.replace("'", "\""))
        new_dic = {'s_a':np.array(dic['s_a']),'n':np.array(dic['n']),'outcome':np.array(dic['outcome'])}
        try:
            curve_optimizer.calc_wl_params(new_dic['s_a'], new_dic['n'], new_dic['outcome'])
            try:
                woehler_params = curve_optimizer.Mali_4p_result
                ratios = curve_optimizer.uncertainty_spread_dict
                df.loc[i,'s_d'] = float(woehler_params['s_d'])
                df.loc[i,'one_t_s'] = float(woehler_params['one_t_s'])
                df.loc[i,'k'] = float(woehler_params['k'])
                df.loc[i,'n_e'] = float(woehler_params['n_e'])

                df.loc[i,'s_d_conf'] = float(ratios['s_d'])
                df.loc[i,'one_t_s_conf'] = float(ratios['one_t_s'])
                df.loc[i,'k_conf'] = float(ratios['k'])
                df.loc[i,'n_e_conf'] = float(ratios['n_e'])
            except AttributeError as a:
                print(f"An error occurred: {a}")
        except ValueError as e:
            print(f"An error occurred: {e}")
    
    return df
```

Processing Outliers

```python
# Remove outliers
def remove_outliers(df,column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column > lower_bound) & (column < upper_bound)]

for col in df.columns:
df.loc[:, col] = remove_outliers(df[col])
```
